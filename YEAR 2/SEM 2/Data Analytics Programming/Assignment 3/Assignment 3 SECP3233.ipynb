You are provided with a rainfall dataset recording monthly measurements from various weather stations across different regions. The dataset includes rainfall (mm), number of rainy days, maximum and minimum temperatures, and whether the station is in an urban or rural area.

Dataset columns:

- `Region`: Name of the region (e.g., North, South, East, West)
- `Year`: Year of measurement
- `Month`: Month (e.g., Jan, Feb, ..., Dec)
- `Station`: Weather station name/ID (e.g., StationA, StationB)
- `Rainfall_mm`: Rainfall amount in millimeters
- `RainyDays`: Number of days in the month with rain
- `MaxTemp_C`: Maximum temperature recorded in that month (°C)
- `MinTemp_C`: Minimum temperature recorded in that month (°C)
- `UrbanRural`: "Urban" or "Rural"

## Part 2: Grouping and Aggregation (30 Marks)

1. For each **Station**, compute: (6 Marks)
    - Total rainfall 
    - Average number of rainy days 
    - Average maximum temperature 
    - Average minimum temperature



```python
#Write your code here

Station=df.groupby(['Station']).agg({
    'Rainfall_mm':'sum',
    'RainyDays':'mean',
    'MaxTemp_C':'mean',
    'MinTemp_C':'mean'})

Station.rename(columns={
    'Rainfall_mm':'Total Rainfall',
    'RainyDays':'Avg Rainy Days',
    'MaxTemp_C':'Avg Max Temp',
    'MinTemp_C':'Avg Min Temp'},inplace=True)

Station
```

2. For each **Region & UrbanRural** combination, calculate: (6 Marks)

    - Average number of rainy days


```python
#Write your code here

combination=df.groupby(['Region','UrbanRural']).agg({
    'Rainfall_mm':'mean',
    'MaxTemp_C':'mean',
    'RainyDays':'mean'})

combination.rename(columns={
    'Rainfall_mm':'mean',
    'MaxTemp_C':'mean',
    'RainyDays':'mean'},inplace=True)

combination
```

3. For each **Month** and **Region**, calculate: (6 Marks)
    - Total rainfall
    - Mean rainfall
    - Minimum and maximum rainfall


```python
#Write your code here

combination2=df.groupby(['Month','Region']).agg({
    'Rainfall_mm':['sum','mean','min','max']})

combination2
```

4. Find the **Region and Month** with the highest recorded rainfall in the dataset. (4 Marks)


```python
# 1. Compute the pivot table (or use your existing)
combination3 = df.groupby(['Region','Month'])['Rainfall_mm'].max().unstack()

# 2. Stack to get a Series with MultiIndex
stacked = combination3.stack()

# 3. Find the (Region, Month) with the highest rainfall
highest_location = stacked.idxmax()
highest_value = stacked.max()

# 4. Print the result clearly
print("============================================")
print(f"🌧️ Highest recorded rainfall was: {highest_value:.2f} mm")
print(f"📍 Occurred in:")
print(f"   ➡ Region: {highest_location[0]}")
print(f"   ➡ Month: {highest_location[1]}")
print("============================================")
```

5. For each **Year** and **UrbanRural** group, calculate: (4 Marks)
    - Average rainfall
    - Minimum rainfall
    - Maximum rainfall



```python
#Write your code here

combination4=df.groupby(['Year','UrbanRural']).agg({
    'Rainfall_mm':['mean','min','max']})

combination4
```

6. For each **Region**, calculate the average temperature range per month (i.e., average of MaxTemp_C - MinTemp_C). (4 Marks)

```python
#Write your code here

# 1️⃣ Compute the temperature range for each record
df['TempRange'] = df['MaxTemp_C'] - df['MinTemp_C']

# 2️⃣ Group by Region and Month, then calculate the average TempRange
avg_temp_range = df.groupby(['Region', 'Month'])['TempRange'].mean()

# 3️⃣ Print results clearly
print("============================================")
print("🌡️ Average Temperature Range per Region and Month")
print(avg_temp_range)
print("============================================")
```
