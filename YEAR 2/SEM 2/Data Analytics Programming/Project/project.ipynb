```python
import pandas as pd

import pandas as pd

# === STEP 1: Load both datasets ===
cars_2025 = pd.read_csv("cars_2025.csv")
fuel_consumption = pd.read_csv("FuelConsumption.csv")

fuel_consumption.info()
fuel_consumption.describe()
fuel_consumption.head()

cars_2025.info()
cars_2025.describe()
cars_2025.head()

print(cars_2025.isnull().sum())

# === STEP 2: Standardize column names (lowercase, no spaces) ===
cars_2025.columns = cars_2025.columns.str.strip().str.lower().str.replace(' ', '_')
fuel_consumption.columns = fuel_consumption.columns.str.strip().str.lower().str.replace(' ', '_')

# === STEP 3: Rename for consistency ===
# 'maker' in cars_2025 becomes 'make'
cars_2025.rename(columns={'maker': 'make'}, inplace=True)

# === STEP 4: Standardize formatting (text uppercase for matching) ===
cars_2025['make'] = cars_2025['make'].str.strip().str.upper()
cars_2025['model'] = cars_2025['model'].str.strip().str.upper()
fuel_consumption['make'] = fuel_consumption['make'].str.strip().str.upper()
fuel_consumption['model'] = fuel_consumption['model'].str.strip().str.upper()

# === STEP 5: Remove outliers from fuel consumption data ===
# Using reasonable cutoffs (e.g., >30L/100km and >600g/km CO2 are outliers)
fuel_cleaned = fuel_consumption[
    (fuel_consumption['fuelconsumption_comb'] < 30) &
    (fuel_consumption['co2emissions'] < 600)
]

# === STEP 6: Optional — show dataset sizes after cleaning ===
print("✅ Data cleaning completed for both dataset respectively.")
print(f"Cleaned cars_2025 shape: {cars_2025.shape}")
print(f"Cleaned fuel_consumption shape: {fuel_cleaned.shape}")

# Merge datasets on make and model to match registrations with emissions info
merged_data = pd.merge(
    cars_2025,
    fuel_cleaned,
    on=['make', 'model'],
    how='inner'
)

# Drop duplicates just in case
merged_data.drop_duplicates(inplace=True)

# Show the shape of the merged dataset
print(f"Merged data shape: {merged_data.shape}")

# Show the first 5 rows to see all columns combined
print(merged_data.head())

# Optional: display the column names to confirm all columns are included
print("Columns in merged data:")
print(merged_data.columns.tolist())

fueltype_summary = merged_data.groupby('fueltype')[['co2emissions', 'enginesize', 'cylinders']].agg(['mean', 'std'])
print(fueltype_summary)
print()

transmission_counts = merged_data.groupby(['fueltype', 'transmission']).size().unstack(fill_value=0)
print(transmission_counts)
print()

# Correlation between emissions and numeric features
correlation_matrix = merged_data[['co2emissions', 'enginesize', 'cylinders', 'fuelconsumption_comb']].corr()
print(correlation_matrix)
print()

# Average emissions grouped by transmission and fuel type
feature_impact = merged_data.groupby(['transmission', 'fueltype'])['co2emissions'].agg(['mean', 'count']).reset_index()
print(feature_impact.head())
```

```python
import pandas as pd
from IPython.display import display, HTML

# Data
data = {
    "Column Name": [
        "make", "model", "enginesize", "cylinders", "transmission",
        "fueltype", "fuelconsumption_comb", "Co2emissions"
    ],
    "Description": [
        "Car brand (e.g. Honda, Toyota)",
        "Car model name",
        "Engine size in litres",
        "Number of engine cylinders",
        "Type of gear system (manual or automatic)",
        "Type of fuel used (e.g. petrol, diesel)",
        "Average fuel usage in L/100km",
        "CO2 emissions in grams per kilometer (g/km)"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Build a left-aligned HTML table manually
html = '<table style="border-collapse: collapse; width: 100%;" border="1">'
html += '<thead><tr><th style="text-align: left;">Column Name</th><th style="text-align: left;">Description</th></tr></thead><tbody>'

for col, desc in zip(df['Column Name'], df['Description']):
    html += f'<tr><td style="text-align: left;">{col}</td><td style="text-align: left;">{desc}</td></tr>'

html += '</tbody></table>'

# Display it
display(HTML(html))
```
